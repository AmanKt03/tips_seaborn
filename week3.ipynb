```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
tips = sns.load_dataset("tips")
```

```python
# 1. Distribution of total bills
plt.figure(figsize=(8, 5))
sns.histplot(tips['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Scatter plot of total bill vs tip
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title('Tip vs Total Bill by Gender')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# 3. Box plot of tips by day
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x="day", y="tip", hue="sex")
plt.title('Tip Distribution by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
# 4. Count plot of smokers vs non-smokers
plt.figure(figsize=(6, 4))
sns.countplot(data=tips, x="smoker", hue="sex")
plt.title('Count of Smokers by Gender')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.show()
```

```python
# 5. Heatmap of correlation
plt.figure(figsize=(6, 5))
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
